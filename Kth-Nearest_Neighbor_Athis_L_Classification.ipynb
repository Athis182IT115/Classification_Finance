```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("fraud.csv")
```

```python
dataset
```

```python
dataset=dataset.drop("isFlaggedFraud",axis=1)
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[["step","amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest","type_CASH_OUT","type_DEBIT","type_PAYMENT","type_TRANSFER"]]
dep=dataset[["isFraud"]]
```

```python
indep
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=1,metric='minkowski',p=2)
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(Y_test,Y_pred)
print(cr)
```

# Accuracy for Kth- Nearest Neighbor Algorithm = 0.98
